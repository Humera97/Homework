```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors")
doc = BeautifulSoup(response.text, 'html.parser')
doc.head()
```

## Part 1

```python
items = doc.find_all('tr', itemtype = 'http://schema.org/Book')
items[1]
```

```python
rows = []
for item in items :
    row = {}
    row['rank'] = item.find('td', class_ = 'number').string
    row['title'] = item.find('td', width= '100%').find('span', itemprop ='name').string
    row['author'] = item.find('a', class_= 'authorName').string
    score_and_vote = item.find('span', class_ = 'smallText uitext').find_all('a')
    row['score'] = score_and_vote[0].string
    row['vote'] = score_and_vote[1].string
    row['rating'] = item.find('span',class_ ='minirating').text
    rows.append(row)

    
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df.to_csv('book_rankings.csv',index = False)
```
