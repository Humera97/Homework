```python
rows =[]
url = ''
for page_num in range(1,5): 
    url = f'http://www.metrolyrics.com/rem-alpage-{page_num}.html'
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    songs = doc.find('table',class_ = 'songs-table compact').find('tbody').find_all('tr')
    #print(songs)
    for song in songs:
        row = {}
        row['title'] = song.find('a').text
        row['url'] = song.find('a')['href']
        row['popularity'] = song.find('span').find('span')['style']
        row['year']= song.find_all('td')[2].text
        rows.append(row)
rows
        
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df.to_csv("metrolyrics_song_title.csv", index=False)
```

```python
df = pd.read_csv("metrolyrics_song_title.csv")
def scrape_page(row):
    url = row['url']
    print("Scraping", url)
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    page = {}
    lyrics = doc.find('div', id = 'lyrics-body-text').find_all('p',class_='verse')
    text =[]
    for lyric in lyrics:
        verse = lyric.text
        text.append(verse)
    page['lyrics'] = text
    return pd.Series(page)
    
```

```python
scraped_df = df.apply(scrape_page, axis=1)
scraped_df
```

```python
df.merge(scraped_df, left_index=True, right_index=True)
```

```python
df.to_csv("metrolyrics_with_lyrics.csv",index= False)
```
